## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

The Refresh / Content Opportunity Scoring task is a Ranking / Scoring problem.

Reason:-
* The Nature of the Decision: Editors face a resource constraint; they
cannot refresh all 22,000+ decaying pages at once. The goal is not to
perform a binary classification (i.e., simply labeling a page as "should be fixed" vs. "should not"), but to prioritize them.

* The Output: We are generating a priority score for each content item that allows an editorial team to sort their backlog. By ranking these items, we ensure that the most high-impact, high-recovery-potential pages are addressed first.

* The Trade-off: While classification could identify which pages are eligible for a refresh, it fails to account for the differential impact of updating those pages. Ranking provides a actionable, ordinal sequence that aligns directly with the practical workflow of a content team.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

An if-statement relies on static thresholds and linear logic, which fail here because content decay is a non-linear, multi-dimensional signal that shifts constantly.

To break it down for your notebook:

* Tangled Signals: Content decline isn't triggered by one factor; it’s a collision of shifting search intent, competitor content updates, technical SEO issues, and seasonal demand. A rule like if traffic < X misses the why, while an ML model can weigh these disparate features (e.g., impression_tier vs position_tier) simultaneously.

* Shifting Baselines: What counts as a "healthy" traffic level or a "normal" decay rate changes per client and per content type. A hand-written rule would require thousands of unique if-statements to account for these variances, whereas a model learns these patterns directly from the data.

* Unpredictable Dynamics: Search engine algorithms evolve; the relationship between a high scroll_rate and future_traffic is not constant. Machine learning adapts to these evolving patterns, whereas an if-statement would become obsolete the moment user behavior or search dynamics change.

In short, a rule-based system learns the definition of decay, but an ML model learns the reality of how that decay manifests in your specific, messy ecosystem.
