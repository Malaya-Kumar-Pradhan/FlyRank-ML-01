```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(5)
```

```python
# Total Content Under Decay
total_decaying = df[df['trend_direction']=='down'].shape[0]
percent_decaying = (total_decaying/df.shape[0])*100

# Severity of Traffic Drops
avg_drop = df[df['trend_direction']=='down']['trend_pct'].mean()

# High-Value Target Space
high_value_targets = df[
    (df['trend_direction']=='down') &
    (df['position_tier'].isin(['striking','page_1'])) &
    (df['impression_tier']=='good')
].shape[0]

print(f"Total Content Under Decay: {total_decaying} rows (or {percent_decaying:.1f}%) are currently flagged with a 'trend_direction' of 'down'. This proves that traffic decay isn't an isolated anomaly; it is the dominant state for the vast majority of our content footprint.\n")
print(f"Average Traffic Drop: The average value within that decaying subset shows a {avg_drop:.1f} traffic drop (trend_pct). Losing roughly half the traffic on three-quarters of our pages represents a massive hit to organic visibility and revenue.\n")
print(f"High-Value Target Space: Out of the decaying pages, {high_value_targets} pages sit in the striking or page_1 tiers for position_tier, while still retaining a good status for impression_tier. These high-value, high-impression pages are actively slipping, making them the perfect high-leverage targets where a timely editorial refresh can yield maximum traffic recovery.")
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

Our analysis will be able to provide observed results, confirming current patterns in content performance and traffic decay using historical data. It will offer directional insights, identifying which clusters of content are trending downward and are most likely to respond to intervention. Ultimately, it provides decision-support, acting as a diagnostic tool for editors to prioritize their workload based on quantified recovery potential.

Conversely, our work will never claim causal proof—we cannot definitively state that an update will trigger recovery, as we are modeling patterns rather than conducting controlled experiments. Furthermore, we are not 'predicting Google'; search algorithms are dynamic and opaque systems, and our model is a simplification of those complex interactions, not a roadmap to manipulating search rankings.
